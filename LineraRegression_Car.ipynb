```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("car driving risk analysis.csv")
df.head(10)
```

```python
plt.scatter(df['speed'],df['risk'])
```

```python
x = df[['speed']]
x.mean()
```

```python
y = df[['risk']]
y.mean()
```

```python
plt.scatter(df['speed'],df['risk'])
plt.scatter(x.mean(),y.mean())
plt.xlabel('Risk')
plt.ylabel('Speed')
plt.title('Car Driving Risk Analysis')
```

```python
from sklearn.linear_model import LinearRegression
```

```python

reg = LinearRegression()
```

```python
reg.fit(x,y)
```

```python
df['Predicted_risk'] = reg.predict(x)
df.head(10)
```

```python
reg.predict([[300]])
```

```python
m = reg.coef_
m
```

```python
c = reg.intercept_
c
```

```python
# y = mx + c
m*300 + c
```

```python
plt.scatter(df['speed'],df['risk'])
plt.scatter(x.mean(),y.mean())
plt.scatter(df['speed'],df['Predicted_risk'])
plt.plot(df['speed'],df['Predicted_risk'])
plt.xlabel('Risk')
plt.ylabel('Speed')
plt.title('Car Driving Risk Analysis')
```

```python
reg.score(x,y)
```

```python
reg.score(df[['speed']],df[['Predicted_risk']])
```

```python
df['Loss'] = df['risk']-df['Predicted_risk']
df.head(10)
```
